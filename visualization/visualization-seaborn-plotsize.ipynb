# Visualization: Adjust Plot Size with Seaborn 

In this notebook, I show how to adjust the plot size with **seaborn**.

## Table-of-Content

* [Load Dataset](#load-dataset)
* [Set parameter *figure.figsize*](#figure-figsize)
* [Create *Matplot* Figure](#matplot-figure)

```python
import pandas as pd
import seaborn as sns
from sklearn import datasets
```

## Load Dataset <a class="anchor" id="load-dataset"></a>

Load **iris dataset**.

```python
# Load iris dataset
iris_data = datasets.load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris["species"] = iris_data.target

#. Replace targets (only for plotting)
iris["species"].replace({0:"setosa", 
                         1:"versicolor", 
                         2:"virginica"}, inplace=True)

iris.head()
```

## Set parameter `figure.figsize` <a class="anchor" id="figure-figsize"></a>

The tuple represents the weight and height of the plot in inches.

```python
# Plot size
sns.set(rc={"figure.figsize": (12,6)})

# Plot iris dataset
ax = sns.scatterplot(data=iris, x="sepal length (cm)", y="petal length (cm)", hue="species")
```

## Create *Matplot* Figure <a class="anchor" id="matplot-figure"></a>

Another option is importing *matplotlib library* and create a figure. You have to pass the parameter *figsize*.

<div class="alert alert-block alert-warning">
    <b>Becareful with option 1 since it affects all cells in the notebook</b>.
</div>

```python
import matplotlib.pyplot as plt

# Creat a figure with `figsize` parameter
plt.figure(figsize=(12,6))

# Plot iris dataset
ax = sns.scatterplot(data=iris, x="sepal length (cm)", y="petal length (cm)", hue="species")
```
