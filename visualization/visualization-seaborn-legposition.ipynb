# Visualization: Set Legend Position with Seaborn

In this notebook, I show how to set the legend position with **seaborn**. <br>
This method can be also applied to other visualization libraries such as **matplotlib**.

## Table-of-Content

* [Load Dataset](#load-dataset)
* [Set Legend Position](#legend-position)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
```

## Load Dataset <a class="anchor" id="load-dataset"></a>

Load **iris dataset**.

```python
# Load iris dataset
iris_data = datasets.load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris["species"] = iris_data.target

#. Replace targets (only for plotting)
iris["species"].replace({0:"setosa", 
                         1:"versicolor", 
                         2:"virginica"}, inplace=True)

iris.head()
```

## Set Legend Position <a class="anchor" id="legend-position"></a>

Put legend at the center of the chart.

```python
# Creat a figure with `figsize` parameter
plt.figure(figsize=(12,6))

# Plot iris dataset with `custom` legend
ax = sns.scatterplot(data=iris, x="sepal length (cm)", y="petal length (cm)", hue="species")
ax.set(title=("Iris dataset"), xlabel="Sepal length", ylabel="Petal length")
ax.legend(loc="upper center", bbox_to_anchor=(0.5,-0.1), ncol=3)
plt.tight_layout();
```
